```python
import pandas as pd
pd.set_option("display.max_columns",10)
print("Original data:\n")
df=pd.read_excel("/Users/student/Downloads/Performance.xlsx")
print(df)
```

```python
f=df["Name"].str.strip()
f.to_excel("/Users/student/Downloads/performance1.xlsx")
```

```python
print("\n Replace Value:")
k=df.fillna(method="pad")
print(k)
print("\n Updated list\n")
k.to_excel("/Users/student/Downloads/Performance.xlsx")
```

```python
print("\n Drop one particular column and its values:\n")
k.drop(["Social"],axis=1,inplace=True)
print(k)
print("\n updated list\n")
k.to_excel("/Users/student/Downloads/Performance.xlsx")
```

```python
print("\n Drop NaN rows:\n")
df=pd.read_excel("/Users/student/Downloads/Performance.xlsx")
print(df)
x=df.dropna()
print(x)
print("\n updated list:\n")
x.to_excel("/Users/student/Downloads/Performance.xlsx")
```
